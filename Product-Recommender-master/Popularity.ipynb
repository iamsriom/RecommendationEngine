```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# we will ignore pandas warning 
import warnings
warnings.filterwarnings('ignore')
```

```python
bigOrdersExport = pd.read_csv('orders_export_1.csv')
productsExport = pd.read_csv('products_export.csv')
```

```python
print(bigOrdersExport.shape)
print(bigOrdersExport.columns.tolist())
bigOrdersExport.head()
```

```python
df_order = bigOrdersExport[['Name', 'Lineitem quantity', 'Lineitem name', 'Lineitem sku']]
print(df_order.shape)
df_order.head()
```

```python
df_order['Name'] = df_order['Name'].str.replace('#', '')
```

```python
df_order.head()
```

```python
#Count of user_id for each unique product as recommendation score 
df_order_grouped = df_order.groupby('Lineitem sku').agg({'Name': 'count'}).reset_index()
df_order_grouped.rename(columns = {'Name': 'score'},inplace=True)
df_order_grouped.head(20)
```

```python
#Sort the products on recommendation score 
df_order_sort = df_order_grouped.sort_values(['score', 'Lineitem sku'], ascending = [0,1]) 
      
#Generate a recommendation rank based upon score 
df_order_sort['rank'] = df_order_sort['score'].rank(ascending=0, method='first') 
          
#Get the top 5 recommendations 
popularity_recommendations = df_order_sort.head(5) 
popularity_recommendations 
```
